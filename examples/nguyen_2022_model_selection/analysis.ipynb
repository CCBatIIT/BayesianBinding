# Nguyen et al. 2022 model selection — analysis

Pairwise Bayes factors among the two-component (2C), racemic-mixture (RM), and enantiomeric-mixture (EM)
models for one dataset, computed by **per-model bridge sampling** (the recommended estimator) with the
**nested warp-BAR** estimator shown alongside as the overlap-gated alternative. See `docs/BAYES_FACTORS.md`.

Run `fit.py` then `bayes_factors.py` first; this notebook only loads the saved summary:

```bash
python examples/nguyen_2022_model_selection/fit.py --full
python examples/nguyen_2022_model_selection/bayes_factors.py
```

````python
import json
from pathlib import Path

import pandas as pd

EXAMPLE_DIR = Path.cwd()
if not (EXAMPLE_DIR / "results").exists():
    candidate = EXAMPLE_DIR / "examples" / "nguyen_2022_model_selection"
    if candidate.exists():
        EXAMPLE_DIR = candidate
summary = json.loads((EXAMPLE_DIR / "results" / "bayes_factors" / "bayes_factor_summary.json").read_text())
print("dataset:", summary["dataset"])

# Per-model log marginal likelihoods (the basis of the recommended estimator).
pd.DataFrame(summary["per_model_log_evidence"]).T
````

## Recommended: per-model bridge Bayes factors

`log10 BF` (complex over simple) with the closed-form asymptotic SE (per-model logZ SEs added in quadrature). Per-model bridge is reliable across the full range of Bayes-factor magnitudes and is chain-consistent by construction.

````python
pd.DataFrame(summary["bridge_bayes_factors"])
````

## Alternative: nested warp-BAR (overlap-gated)

The nested between-model estimator with warp-BAR and its Bennett overlap. Where the two posteriors barely overlap — the large 2C-vs-mixture factors — the overlap is low and the estimate is flagged `reliable=False`, deferring to the per-model bridge above (the bias the methods note describes).

````python
pd.DataFrame(summary["nested_warp_bar_bayes_factors"])
````

**Takeaway.** The mixture models (RM/EM) decisively beat the two-component model, and RM vs EM distinguishes a racemic from a non-racemic titrant. The per-model bridge gives trustworthy factors for every pair; the nested warp-BAR agrees where overlap is good and is flagged where it is not.
